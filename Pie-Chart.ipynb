# Introduction of Pie-Chart

```python
import pandas as pd   #import pandas 
import numpy as np   # import numpy
import matplotlib.pyplot as plt # Import matplotlib from matplotlib import pyplot.
```

```python
df = pd.read_csv("TRAIN (1).csv") # Import data
df.head()
```

```python
Location_type = df.Location_Type.value_counts().reset_index()
Location_type
```

```python
location = Location_type['index']
count = Location_type['Location_Type']
```

- To start build the pie chart first we have to get values that we are going to import on pie-chart.
- later we have to get labels for particular pie.
- count is values for pie and location is labels for pie.

```python
plt.figure(figsize = (20,8))
plt.pie(count,labels = location)
plt.title('pie-chart of count of all locations')
```

- now as we already plot graph using "plt.pie()" function, you can see there are few parameters are present in it.
- Wedgeprops used to modify edges of each pie. In following graph you can see as we mention colors to edge that is showing in the graph.

```python
plt.figure(figsize = (20,8))
plt.pie(count,labels = location,wedgeprops = {'edgecolor':'black'})
plt.title('pie-chart of count of all locations')
```

- change colors of pie we can mention list of colours and that list we can pass in pie function.

```python
colors = ['black','purple','yellow','blue','grey']

plt.figure(figsize = (20,8))
plt.pie(count,labels = location,wedgeprops = {'edgecolor':'black'},colors = colors)
plt.title('pie-chart of count of all locations')
```

- Sometimes we have to show one pie as important feature, to show that we have explode parameter in pie that help us to show particular pie as separate from others.

```python
explode = [0,0.2,0,0,0]

plt.figure(figsize = (20,8))
plt.pie(count,labels = location,wedgeprops = {'edgecolor':'black'},explode = explode,shadow = True)
plt.title('pie-chart of count of all locations')
```

- there is another parameter which used to start pie from particular degree that parameter called "startangle".

```python
plt.figure(figsize = (20,8))
plt.pie(count,labels = location,wedgeprops = {'edgecolor':'black'},explode = explode,shadow = True,startangle = 150)
plt.title('pie-chart of count of all locations')
```

- For mentioning percentage to each of pie use  function called "autopct"and mention following format that will help you to show percentage of particular pie. 

```python
plt.figure(figsize = (20,8))
plt.pie(count,labels = location,wedgeprops = {'edgecolor':'black'},explode = explode,shadow = True,startangle = 150,autopct = '%1.1f%%')
plt.title('pie-chart of count of all locations')
```
