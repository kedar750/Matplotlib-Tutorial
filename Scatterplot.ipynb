# Scatter plot Tutorial

# 1.Import Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# 2.Import Data

```python
df = pd.read_csv('TRAIN (1).csv')
df.head()
```

```python
region_id = df[df['Region_Code']=='R1']
region_id
```

- for showing scatter plot Import x data and y data.
- 'c' is used to give color for dots.
- 'marker' is used to changes shpae of dots.

```python
plt.figure(figsize = (21,9))
plt.xlabel('Store_ID')
plt.ylabel('Order per store')
plt.title('Graph for orders per store')
plt.scatter(x = region_id['Store_id'],y = region_id['#Order' ],c = 'purple',marker = '*')
```

- 'Edgecolor' is used to show different color to the edge.
- 'linewidth'is used to define width to edge.

```python
plt.figure(figsize = (21,9))
plt.xlabel('Store_ID')
plt.ylabel('Order per store')
plt.title('Graph for orders per store')
plt.scatter(x = region_id['Store_id'],y = region_id['#Order' ],c = 'black',edgecolor = 'yellow',linewidth = 1,alpha = 0.75)
```

```python
plt.figure(figsize = (21,9))
plt.xlabel('Store_ID')
plt.ylabel('Order per store')
plt.title('Graph for orders per store')
plt.scatter(x = region_id['Store_id'],y = region_id['Sales' ],c = region_id['#Order'],cmap = 'summer',edgecolor = 'pink',linewidth = 1,alpha = 0.75)

cbar = plt.colorbar()
cbar.set_label('Orders sales ratio')
```
